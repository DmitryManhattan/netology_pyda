**Уровень 0:**
Даны вектора x=(1,1), y=(2, 0) и z = (0,2) <br>
Надо: <br>
Изобразить вектора на экране <br>




```python

import numpy as np
import matplotlib.pyplot as plt

plt.figure()

x = np.array([1,1])
y = np.array([2,0])
z = np.array([0,2])

plt.plot([0, x[0]], [0, x[1]], linestyle='-', color = 'blue')
plt.plot([0, y[0]], [0, y[1]], linestyle='-', color='orange')
plt.plot([0, z[0]], [0, z[1]], linestyle='-', color = 'green')

plt.legend(['vector X', 'vector Y', ' vector z'])

plt.grid()
plt.show()
```

Изобразить точку x + y + z <br>

```python
plt.figure()

x = np.array([1,1])
y = np.array([2,0])
z = np.array([0,2])
vector_sum = x + y + z

plt.plot([0, y[0]], [0, y[1]], linestyle='-', color='orange')
plt.plot([y[0], (y[0] + x[0])], [y[1], (y[1] + x[1])], linestyle='-', color = 'blue')
plt.plot([(y[0] + x[0]), (y[0] + x[0] + z[0])], [(y[1] + x[1]), (y[1] + x[1] + z[1])], linestyle='-', color = 'green')
plt.plot([0, vector_sum[0]], [0, vector_sum[1]], linestyle='-', color = 'red')

plt.legend(['vector X', 'vector Y', 'vector z', 'X + Y + Z'])

plt.grid()
plt.show()
```

```python
plt.figure()

x = np.array([1,1])
y = np.array([2,0])
z = np.array([0,2])
vector_sum = x + y + z

plt.plot(x[0], x[1], 'o')
plt.plot(y[0], y[1], 'o')
plt.plot(z[0], z[1], 'o')
plt.plot(vector_sum[0], vector_sum[1], 'o')

plt.grid()
plt.show()
```

Найти угол между векторами x и y <br>

```python
rad = np.arccos (  
    np.dot(x,y) / ( (np.sqrt(x[0]**2 + x[1]**2)) * (np.sqrt(y[0]**2 + y[1]**2)) )  
            )
```

```python
np.degrees(rad)
```
