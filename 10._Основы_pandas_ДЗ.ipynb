Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data_power = pd.read_csv('power.csv')
data_power.head()
```

```python
filtered_data_power = data_power[ (data_power.quantity >= 0) & ((data_power.category == 4) | (data_power.category == 12) | (data_power.category == 21)) & ((data_power.year >= 2005) & (data_power.year <= 2010)) & ((data_power.country == 'Lithuania') | (data_power.country == 'Latvia') | (data_power.country == 'Estonia') )]

filtered_data_power.groupby('country').quantity.sum().reset_index()
```
